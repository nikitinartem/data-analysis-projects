```python
import pandas as pd
```

```python
data = pd.DataFrame({
    'date': pd.date_range(start='2024-01-01', periods=14),
    'category': ['A', 'B'] * 7,
    'sales': [120, 80, 150, 90, 130, 100, 170, 110, 160, 105, 180, 115, 175, 120]
})

data
```

```python
data.isnull().sum()
```

```python
daily_sales = data.groupby('date')['sales'].sum().reset_index()

daily_sales
```

```python
daily_sales['rolling_mean_7d'] = daily_sales['sales'].rolling(window=7).mean()

daily_sales
```
